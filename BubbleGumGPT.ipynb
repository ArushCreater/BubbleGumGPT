```python
from transformers import AutoTokenizer
import json

# Load Reddit dataset
with open("reddit_dataset.json", "r") as f:
    data = json.load(f)

# Initialize tokenizer (smallest model for now)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Convert text to token IDs
def tokenize_data(data):
    tokenized = []
    for entry in data:
        input_tokens = tokenizer(entry["input"], padding="max_length", truncation=True, max_length=50, return_tensors="pt")["input_ids"].squeeze(0)
        output_tokens = tokenizer(entry["output"], padding="max_length", truncation=True, max_length=50, return_tensors="pt")["input_ids"].squeeze(0)
        tokenized.append((input_tokens, output_tokens))
    return tokenized

tokenized_data = tokenize_data(data)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Transformer-based Chatbot Model
class ChatbotTransformer(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, num_heads=8, num_layers=6):
        super(ChatbotTransformer, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads), num_layers=num_layers
        )
        self.decoder = nn.Linear(embed_dim, vocab_size)

    def forward(self, src):
        embedded = self.embedding(src)
        encoded = self.encoder(embedded)
        return self.decoder(encoded)

# Load tokenizer vocab size
vocab_size = tokenizer.vocab_size

# Initialize model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Use GPU if available
model = ChatbotTransformer(vocab_size).to(device)
```

```python
class ChatDataset(Dataset):
    def __init__(self, tokenized_data):
        self.data = tokenized_data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

# Create DataLoader
dataset = ChatDataset(tokenized_data)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)
```

```python
import torch
import torch.optim as optim
import torch.nn as nn

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=5e-4)

# Training loop for 1 epoch
for epoch in range(1):  # Train for 1 epoch (you can increase this later)
    model.train()  # Ensure the model is in training mode
    total_loss = 0

    for batch in dataloader:  # Assuming dataloader is already defined
        # Check the structure of the batch
        print(batch)  # Optional: print batch structure for debugging

        # Access the input tokens and output tokens properly
        input_tokens = batch[0].to(device)  # Assuming first element is input_ids
        output_tokens = batch[1].to(device)  # Assuming second element is labels

        optimizer.zero_grad()  # Clear previous gradients

        # Forward pass
        predictions = model(input_tokens)  # Make predictions

        # Calculate the loss
        loss = criterion(predictions.view(-1, vocab_size), output_tokens.view(-1))  # Cross-entropy loss
        total_loss += loss.item()

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

    # Print loss after every epoch
    print(f"Epoch {epoch+1}, Loss: {total_loss / len(dataloader):.4f}")

# Save the trained model
torch.save(model.state_dict(), "reddit_chatbot.pth")
print("Training complete. Model saved!")
```

```python
# Load trained model
model.load_state_dict(torch.load("reddit_chatbot.pth", map_location=device))
model.eval()

def generate_response(input_text):
    input_tokens = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)["input_ids"].to(device)

    with torch.no_grad():
        # Get the model output
        model_output = model(input_tokens)

    # Extract the logits (first element of model_output)
    logits = model_output[0]  # Shape: (batch_size, sequence_length, vocab_size)
    print("Logits:", logits)  # Debugging line to check logits

    # Process the logits for the last token
    predicted_token_ids = logits.argmax(dim=-1)  # Get the token ID with the highest probability
    print("Predicted Token IDs:", predicted_token_ids)  # Debugging line to check predicted token IDs

    # Convert the token IDs to a list of tokens
    predicted_token_ids = predicted_token_ids.squeeze().cpu().numpy()  # Remove batch dimension

    # Decode the token IDs into text
    response = tokenizer.decode(predicted_token_ids, skip_special_tokens=True)

    # Debugging to check if the response is empty
    print("Decoded Response:", response)

    return response

# Chat with the bot
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break
    print("Bot:", generate_response(user_input))
```
